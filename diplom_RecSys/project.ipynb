## Статистический анализ данных

Нулевая гипотеза (H0): средний рейтинг женщин и мужчин не отличается.

Альтернативная (H1): средний рейтинг женщин и мужчин отличается.

```python
sns.histplot(merged[merged['gender'] == 'M']['rating'])
```

```python
from scipy import stats

alpha = 0.05 # уровень значимости

merged = data.merge(user_df, on='user_id')
ratings_male = merged[merged['gender'] == 'M']['rating']
ratings_female = merged[merged['gender'] == 'F']['rating']

shapiro_result_a = stats.shapiro(ratings_male)
shapiro_result_b = stats.shapiro(ratings_female)
print('p-value группы А', round(shapiro_result_a.pvalue, 2))
print('p-value группы B', round(shapiro_result_b.pvalue, 2))

# сравниваем полученное p-value для группы А с уровнем значимости
if shapiro_result_a.pvalue <= alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной. Распределение в группе А отлично от нормального")
else:
    print("Принимаем нулевую гипотезу. Распределение в группе А является нормальным")

# сравниваем полученное p-value для группы B с уровнем значимости
if shapiro_result_b.pvalue <= alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной. Распределение в группе B отлично от нормального")
else:
    print("Принимаем нулевую гипотезу. Распределение в группе B является нормальным")
```

```python
# вычисляем результат теста Манна — Уитни для выборок
results = stats.mannwhitneyu(
    x=ratings_male,
    y=ratings_female,
    alternative='two-sided'
)
print('p-value:', round(results.pvalue, 2))

# сравниваем полученное p-value с уровнем значимости
if results.pvalue <= alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("У нас нет оснований отвергнуть нулевую гипотезу")
```

```python
results = stats.mannwhitneyu(
    x=ratings_male,
    y=ratings_female,
    alternative='less'
)
print('p-value:', round(results.pvalue, 2))

# сравниваем полученное p-value с уровнем значимости
if results.pvalue <= alpha:
    print("Отвергаем нулевую гипотезу в пользу альтернативной")
else:
    print("У нас нет оснований отвергнуть нулевую гипотезу")
```

*Вывод:* 
- Средний рейтинг мужчин и женщин  статистически отличается
- средний рейтинг женщин выше

Нулевая гипотеза (H0): средние значения рейтингов всех групп (профессий) равны.

Альтернативная (H1): средние значения рейтингов всех групп (профессий) неравны.

```python
professions = merged['occupation'].unique()

# Результаты проверки
normality_results = {}

for prof in professions:
    ratings = merged.loc[merged['occupation'] == prof, 'rating']
    shapiro_result = stats.shapiro(ratings)
    normality_results[prof] = shapiro_result.pvalue

# Отобразим результаты
for prof, result in normality_results.items():
    print('p-value:', round(shapiro_result.pvalue, 2))

# сравниваем полученное p-value с уровнем значимости
    if shapiro_result.pvalue <= alpha:
       print("Отвергаем нулевую гипотезу в пользу альтернативной")
    else:
       print("У нас нет оснований отвергнуть нулевую гипотезу")
```

```python
professions = merged['occupation'].unique()
groups = [merged.loc[merged['occupation'] == prof, 'rating'] for prof in professions]

kruskal_result = stats.kruskal(*groups)

print('p-value = {:.3f}'.format(kruskal_result.pvalue))

# интерпретируем результат
if kruskal_result.pvalue <= alpha:
    print('p-значение меньше, чем заданный уровень значимости {:.2f}. Отвергаем нулевую гипотезу.'.format(alpha))
else:
    print('p-значение больше, чем заданный уровень значимости {:.2f}. У нас нет оснований отвергнуть нулевую гипотезу.'.format(alpha))
```

*Вывод:* 
- средние значения рейтингов всех групп (профессий) неравны.

#### Content-Based рекомендации по жанрам

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# item_df - DataFrame с информацией о фильмах и бинарными жанрами в колонках с 6-го столбца
genre_columns = item_df.columns[5:]

# Кодируем жанры с помощью OneHotEncoder, каждая строка-фильм, в столбцах наличие жанра у фильма
encoder = OneHotEncoder(sparse_output=False)
genres_encoded = encoder.fit_transform(item_df[genre_columns])

# Создаём профиль пользователя — средневзвешенный жанровой вектор по его оценкам
def user_profile(user_id):
    user_ratings = data[data['user_id'] == user_id].merge(item_df, on='item_id')
    user_genres = genres_encoded[user_ratings.index]
    weights = user_ratings['rating'].values #веса для каждого фильма, влияющие на среднее — чем выше рейтинг, тем больше вклад этого фильма
    profile = np.average(user_genres, axis=0, weights=weights)
    return profile

# Рекомендуем фильмы по косинусному сходству жанров к профилю пользователя
def recommend_content(user_id, top_n=10):
    from sklearn.metrics.pairwise import cosine_similarity
    profile = user_profile(user_id).reshape(1, -1)
    similarities = cosine_similarity(genres_encoded, profile).flatten()
    recommended_indices = np.argsort(similarities)[::-1] #argsort возвращает индексы элементов массива similarities, которые отсортировали бы этот массив в порядке возрастания
    recommended_items = item_df.iloc[recommended_indices]['title']
    return recommended_items.head(top_n).tolist()

user_id = 1
rec_content = recommend_content(user_id)
print(f"Content-based рекомендации для пользователя {user_id}:", rec_content)
```

#### Коллаборативная фильтрация с implicit (ALS)

```python
import scipy.sparse as sparse
import implicit
import numpy as np

# Формируем разреженную матрицу user-item с бинарными взаимодействиями (1 если есть рейтинг)
user_ids = data['user_id'].values
item_ids = data['item_id'].values
values = np.ones(len(data))  # для неявных данных считаем просто наличие рейтинга

user_item_matrix = sparse.coo_matrix((values, (user_ids, item_ids)))

# Обучаем ALS модель
model_als = implicit.als.AlternatingLeastSquares(factors=50, regularization=0.01, iterations=20)
model_als.fit(user_item_matrix.T)  # транспонируем под требования implicit

# Рекомендации для пользователя
user_to_recommend = 1
recommended = model_als.recommend(user_to_recommend, user_item_matrix.tocsr(), N=10)
print(f"Implicit ALS рекомендации для пользователя {user_to_recommend}:", recommended)
```
